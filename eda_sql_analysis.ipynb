```python
import pandas as pd
```

```python

df = pd.read_excel("data.xlsx")

df.head()
```

```python
df.shape
df.info()
df.describe(include='all')
```

```python
df.to_csv("data.csv", index=False)
```

```python
df_clean = df.copy()
```

```python
df_clean.isnull().sum()
```

```python
df_clean['total_charges'] = pd.to_numeric(
    df_clean['total_charges'],
    errors='coerce'
)
```

```python
df_clean.info()
```

```python
df_clean.duplicated().sum()
df_clean.drop_duplicates(inplace=True)
```

```python
df_clean['churn_value'].value_counts(normalize=True) * 100
```

```python
df_clean['total_charges'].isnull().sum()
```

```python
df_clean['total_charges'].fillna(0, inplace=True)
```

```python
columns_to_drop = [
    'customerid',
    'count',
    'zip_code',
    'lat_long',
    'latitude',
    'longitude'
]

df_clean.drop(columns=columns_to_drop, inplace=True)
```

```python
binary_cols = [
    'senior_citizen', 'partner', 'dependents', 'phone_service',
    'paperless_billing'
]

for col in binary_cols:
    df_clean[col] = df_clean[col].map({'Yes': 1, 'No': 0})
```

```python
service_cols = [
    'multiple_lines', 'online_security', 'online_backup',
    'device_protection', 'tech_support',
    'streaming_tv', 'streaming_movies'
]

for col in service_cols:
    df_clean[col] = df_clean[col].replace(
        {'No internet service': 'No', 'No phone service': 'No'}
    )
```

```python
df_clean['tenure_group'] = pd.cut(
    df_clean['tenure_months'],
    bins=[0, 12, 24, 36, 48, 60, 100],
    labels=['0–1 year', '1–2 years', '2–3 years', '3–4 years', '4–5 years', '5+ years']
)
```

```python
df_clean['churn_reason'] = df_clean['churn_reason'].fillna('Not Churned')
```

```python
df_clean.info()
df_clean.isnull().sum()
```

```python
df_clean.to_csv("telco_churn_cleaned.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

churn_rate = df_clean['churn_value'].mean() * 100
print(f"Overall Churn Rate: {churn_rate:.2f}%")
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='churn_label', y='tenure_months', data=df_clean)
plt.title("Tenure vs Churn")
plt.show()
```

```python
tenure_churn = (
    df_clean
    .groupby('tenure_group')['churn_value']
    .mean()
    .sort_values(ascending=False)
)

tenure_churn.plot(kind='bar', figsize=(8,5), title="Churn Rate by Tenure Group")
plt.ylabel("Churn Rate")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(
    data=df_clean,
    x='monthly_charges',
    hue='churn_label',
    bins=30,
    kde=True
)
plt.title("Monthly Charges Distribution by Churn")
plt.show()
```

```python
contract_churn = (
    df_clean
    .groupby('contract')['churn_value']
    .mean()
    .sort_values(ascending=False)
)

contract_churn.plot(kind='bar', figsize=(7,5), title="Churn Rate by Contract Type")
plt.ylabel("Churn Rate")
plt.show()
```

```python
payment_churn = (
    df_clean
    .groupby('payment_method')['churn_value']
    .mean()
    .sort_values(ascending=False)
)

payment_churn.plot(kind='bar', figsize=(9,5), title="Churn Rate by Payment Method")
plt.ylabel("Churn Rate")
plt.show()
```

```python
internet_churn = (
    df_clean
    .groupby('internet_service')['churn_value']
    .mean()
    .sort_values(ascending=False)
)

internet_churn.plot(kind='bar', figsize=(7,5), title="Churn Rate by Internet Service")
plt.ylabel("Churn Rate")
plt.show()
```

```python
!pip install pandasql
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
pysqldf("""
SELECT 
    ROUND(AVG(churn_value) * 100, 2) AS churn_rate_percent
FROM df_clean;
""")
```

```python
pysqldf("""
SELECT 
    contract,
    COUNT(*) AS total_customers,
    SUM(churn_value) AS churned_customers,
    ROUND(AVG(churn_value) * 100, 2) AS churn_rate_percent
FROM df_clean
GROUP BY contract
ORDER BY churn_rate_percent DESC;
""")
```

```python
pysqldf("""
SELECT 
    payment_method,
    COUNT(*) AS total_customers,
    ROUND(AVG(churn_value) * 100, 2) AS churn_rate_percent
FROM df_clean
GROUP BY payment_method
ORDER BY churn_rate_percent DESC;
""")
```

```python
pysqldf("""
SELECT 
    internet_service,
    COUNT(*) AS total_customers,
    ROUND(AVG(churn_value) * 100, 2) AS churn_rate_percent
FROM df_clean
GROUP BY internet_service
ORDER BY churn_rate_percent DESC;
""")
```

```python
pysqldf("""
SELECT
    CASE 
        WHEN tenure_months < 12 THEN '0-1 year'
        WHEN tenure_months BETWEEN 12 AND 24 THEN '1-2 years'
        WHEN tenure_months BETWEEN 25 AND 48 THEN '2-4 years'
        ELSE '4+ years'
    END AS tenure_group,
    COUNT(*) AS total_customers,
    ROUND(AVG(churn_value) * 100, 2) AS churn_rate_percent
FROM df_clean
GROUP BY tenure_group
ORDER BY churn_rate_percent DESC;
""")
```

```python
pysqldf("""
SELECT
    CASE 
        WHEN cltv >= (SELECT AVG(cltv) FROM df_clean) THEN 'High Value'
        ELSE 'Low Value'
    END AS customer_segment,
    COUNT(*) AS total_customers,
    ROUND(AVG(churn_value) * 100, 2) AS churn_rate_percent
FROM df_clean
GROUP BY customer_segment;
""")
```
